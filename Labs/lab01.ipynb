### We approximate the integral of the function cos(x) in the range [-pi, pi] using Riemann Midpoint Summation. The areas of the rectngles with midpoints are summed to approximate the integral. The exact integral of cos(x) in the range [-pi, pi] is zero. Thereby as we increase the interval 'n' the approximate value approaches zero.

The function riemann returns the total sum, the midpoints (x-values), and the heights (y-value) which are used to plot a bar graph for the summation.

The function calcErrs is used to calculate the absolute error given by the formula 'Error = | exact - approx |', over the range of intervals from 1 to 100, and then finally using the errors to plot the graph of erros using logarithmic scale.


```python
def riemann(a, b, f, n):
    sum = 0.0
    dx = (b - a) / n
    midpoints = []
    heights = []
    for x0 in range(n):
        pt = ((a + x0 * dx) + (a + (x0 + 1) * dx)) / 2 # Calculating midpoint
        midpoints.append(pt)
        heights.append(f(pt))
        sum = sum + f(pt)
    return sum, np.array(midpoints), np.array(heights)
```

```python
a = -np.pi
b = np.pi
n = 10
dx = (b - a) / n
_, midpoints, heights = riemann(a, b, np.cos, n)

print(_)

plt.plot(midpoints, heights)
plt.plot(midpoints, heights,'b.',markersize=10)
plt.bar(midpoints, heights ,width=dx,alpha=0.2,edgecolor='b')
plt.title('Midpoint Riemann Sum, N = {}'.format(n))
```

```python
a = -np.pi
b = np.pi
n = 100
dx = (b - a) / n
_, midpoints, heights = riemann(a, b, np.cos, n)

plt.plot(midpoints, heights)
plt.plot(midpoints, heights,'b.',markersize=10)
plt.bar(midpoints, heights ,width=dx,alpha=0.2,edgecolor='b')
plt.title('Midpoint Riemann Sum, N = {}'.format(n))
```

```python
def calcErrs(f):
    n_vals = np.linspace(1, 100, 100)

    errs = []
    for n in n_vals:
        approx, _, _ = riemann(-np.pi, np.pi, f, int(n))
        err = np.abs(approx - 0) # Error = | Exact Integral - Approx |
        errs.append(err)

    return np.array(errs)
```

```python

errs = calcErrs(np.cos)
n_vals = np.linspace(1, 100, 100)

plt.xlabel('x')
plt.ylabel('f(x) = cos(x)')
plt.yscale("log")
plt.xscale("log")
plt.plot(n_vals, errs, marker="o", color="green")
```
